```python
import matplotlib.pyplot as plt
import numpy as np

data = [7, 9, 17, 45, 48, 51, 52, 63, 65, 72, 130]

Q1 = np.percentile(data, 25)  
Q3 = np.percentile(data, 75)  

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

plt.boxplot(data, vert=False, patch_artist=True)

plt.axvline(lower_limit, color='red', linestyle='--', label=f'Lower Limit: {lower_limit:.2f}')
plt.axvline(upper_limit, color='green', linestyle='--', label=f'Upper Limit: {upper_limit:.2f}')

for point in data:
    plt.text(point, 1, str(point))


plt.legend()
plt.show()
```
